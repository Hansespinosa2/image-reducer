```python
import sklearn
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
for root, _, file_paths in os.walk('images'):
    the_file_paths = [os.path.join(root,path) for path in file_paths]

the_file_paths
```

```python
from sklearn.datasets import load_sample_images
images = []
for path in the_file_paths:
    iwa = plt.imread(path)
    iwa = iwa[:240,:240,:3]
    plt.imshow(iwa)
    plt.axis('off')
    plt.show()
    images.append(iwa)

images = tf.keras.layers.CenterCrop(height=240,width=240)(images)
images
```

```python
images.shape
```

```python
tf.random.set_seed(42)

max_pool = tf.keras.layers.MaxPool2D(pool_size=4)

output = max_pool(images)
```

```python
# extra code – this cells shows what max pooling with stride = 2 looks like

import matplotlib as mpl



fig = plt.figure(figsize=(12, 8))

ax1 = fig.add_subplot(111)
ax1.set_title("Input")
ax1.imshow(images[0])  # plot the 1st image
ax1.axis("off")
ax2 = fig.add_subplot(211)
ax2.set_title("Output")
ax2.imshow(output[0])  # plot the output for the 1st image
ax2.axis("off")
plt.show()
```
